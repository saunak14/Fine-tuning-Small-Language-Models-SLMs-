### Ollama wrapper to import SLMs easily

```python
import ollama
```

#### Importing CSV as a regular file

```python
import csv
with open('employees.csv', 'r') as f:
    all_employees = f.read()
print(all_employees)
```

### Prompt engineering
#### Giving the SLM model all data

```python
prompt = f"""
I will give you a list of employees where each record is present in a newline. The first line contains the following headers: Name, Designation and EmployeeID.\
The list is present inside the backticks \
How many employees have Designation as 'HR Manager'? \

`
{all_employees}
`

"""
print(prompt)
```

```python
response = ollama.chat(model='phi3', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
print(response['message']['content'])
```
